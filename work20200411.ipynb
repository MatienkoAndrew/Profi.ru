Напиши функцию, которая удалит дубликаты в списке. Список не отсортирован. Необходимо сохранить порядок сортировки оригинального списка.

Примеры:
    
- [1, 2, 3, 1] → [1, 2, 3]

- [1, 3, 2, 1, 5, 3, 5, 1, 4] → [1, 3, 2, 5, 4]

Какая асимптотическая сложность у этой функций?


```python
array = [1,2,3,1]
array_1 = [1, 3, 2, 1, 5, 3, 5, 1, 4]
```

```python
def unique(ar):
    ar_new = []
    for value in ar:
        if value not in ar_new:
            ar_new.append(value)
    return ar_new
```

```python
unique(array)
```

```python
unique(array_1)
```

### Сложность функции unique = O($n!$), так как мы пробегаемся по массиву $n$ раз в цикле $for\ value\ in\ ar$. И $(n-1)*(n-2)*(n-3)*...*1$ при проверке условия в операторе $not\ in$

###  Решить задачу

Подкинули монету N раз. Кол-во случаев, когда выпал орёл, на 10% больше, чем кол-во случаев, когда выпала решка. При каком N мы можем сказать, что монета «нечестная» (орёл и решка выпадают с разной вероятностью)?


## Будем решать через критерий согласия хи-квадрат (Пирсона)

- $H_0$ - нулевая гипотеза: монетка обычная

- $H_1$ - альтернативная гипотеза: монетка "нечестная"

### Формула критерия согласия Пирсона

,где:

- $n$ - количество подбрасываний монеты

- $p_i$ - статистическая вероятность выпадения орла/решки. $p_i$ = $\frac{1}{2}$

- $p^*_i$ - эмпирическая вероятность выпадения орла/решки

- $k$ - все варианты. $k$=2 : либо орел, либо решка

- $r$ - число степеней свободы: $r = k - s$. 

$s$ = 1, так как одно условие: $\sum^k_{i=1}{p^*_i=1}$

$r = 2 - 1 = 1$

$\alpha$ - уровень значимости. 

Возьмем уровень значимости $\alpha = 0.05$. То есть если наше значение хи-квадрат будет больше $3.8$ (по таблице), то вероятность того, что монета "нечестная" будет 95%

![image](https://user-images.githubusercontent.com/29499863/98159464-b6ed0080-1eed-11eb-9030-c14f13083660.png)

Выводя $n$ из формулы Пирсона получаем:

$n = \frac{\chi^2}{2 \cdot \sum^k_i=1(p^*_i - \frac{1}{2})^2}$

```python
import numpy as np
chi = 3.9
p_star = np.array([0.5238095238095238, 0.47619047619047616])
p = np.array([1/2, 1/2])
n = round(chi / (2 * sum((p_star - p)**2)))
```

```python
n
```

### Ответ: при 1720 подбрасываниях можно заметить, что монета "нечестная"
